```python
# Feature Engineering: Creating new features
df["Income/Age"] = df["Income"] / df["Age"]
df["Experience/Age"] = df["Experience"] / df["Age"]
df["Job_House_Ratio"] = df["CURRENT_JOB_YRS"] / (df["CURRENT_HOUSE_YRS"] + 1)
df["Income_Experience_Ratio"] = df["Income"] / (df["Experience"] + 1)
df["Total_Years"] = df["CURRENT_JOB_YRS"] + df["CURRENT_HOUSE_YRS"]
df["Debt_Income_Ratio"] = df["Income"] / (df["CURRENT_HOUSE_YRS"] + 1)
df["Income_per_Job_Year"] = df["Income"] / (df["CURRENT_JOB_YRS"] + 1)
df["Experience_per_House_Year"] = df["Experience"] / (df["CURRENT_HOUSE_YRS"] + 1)
df["Total_Assets"] = df["Income"] * df["CURRENT_HOUSE_YRS"]
df["Experience_Age_Ratio"] = df["Experience"] / df["Age"]
df["Income_Group"] = pd.qcut(df["Income"], q=4, labels=["Low", "Medium", "High", "Very High"])
```

```python
# Define models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier


models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Decision Tree": DecisionTreeClassifier(),
    "KNN": KNeighborsClassifier(),
}

```

```python
# Number of iterations
num_iterations = 5
```

```python
# Train-test split and model evaluation over multiple runs
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score
for i in range(num_iterations):
    print(f"\nIteration {i+1}:")
    X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=i, stratify=y)
    
    for name, model in models.items():
        pipeline = Pipeline([
            ("classifier", model)
        ])
        
        # Train model
        pipeline.fit(X_train, y_train)
        
        # Predictions
        y_pred = pipeline.predict(X_test)
        y_prob = pipeline.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None
        
        # Print performance metrics
        print(f"\n{name} Model Performance:")
        print("Accuracy:", accuracy_score(y_test, y_pred))
        print("Classification Report:\n", classification_report(y_test, y_pred))
        if y_prob is not None:
            print("AUC Score:", roc_auc_score(y_test, y_prob))
```

```python
from flask import Flask, request, jsonify, render_template_string
import joblib
import pandas as pd
import threading

# Initialize Flask app
app = Flask(__name__)

# Load the trained model
model = joblib.load("loan_risk_model.pkl")

# Define categorical mappings
category_mappings = {
    "Car_Ownership": {"no": 0, "yes": 1},
    "Married/Single ": {"single": 0, "married": 1},
    "House_Ownership": {"rented": 0, "owned": 1, "norent_noown": 2}
}

# HTML page
HTML_PAGE = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Loan Risk Prediction</title>
    <style>
        body { font-family: Arial, sans-serif; text-align: center; margin: 50px; }
        .container { max-width: 500px; margin: auto; padding: 20px; border: 1px solid #ddd; border-radius: 10px; }
        input, select { width: 100%; padding: 10px; margin: 10px 0; }
        button { padding: 10px 20px; cursor: pointer; background: blue; color: white; border: none; }
        #result { font-size: 20px; margin-top: 20px; }
    </style>
</head>
<body>
    <div class="container">
        <h2>Loan Risk Prediction</h2>
        <form id="loanForm">
            <input type="number" id="income_experience" placeholder="Income Experience Ratio" required>
            <input type="number" id="income_per_job" placeholder="Income per Job Year" required>
            <input type="number" id="age" placeholder="Age" required>
            <input type="number" id="exp_per_house" placeholder="Experience per House Year" required>
            <input type="number" id="total_years" placeholder="Total Years" required>
            <input type="number" id="current_job" placeholder="Current Job Years" required>
            <input type="number" id="experience" placeholder="Experience" required>
            
            <select id="car_ownership">
                <option value="no">No Car</option>
                <option value="yes">Owns Car</option>
            </select>
            
            <select id="marital_status">
                <option value="single">Single</option>
                <option value="married">Married</option>
            </select>
            
            <select id="house_ownership">
                <option value="rented">Rented</option>
                <option value="owned">Owned</option>
                <option value="norent_noown">No Rent No Own</option>
            </select>
            
            <button type="button" onclick="predictLoan()">Predict</button>
        </form>
        
        <p id="result"></p>
    </div>

    <script>
        function predictLoan() {
            let data = {
                "Income_Experience_Ratio": parseFloat(document.getElementById("income_experience").value),
                "Income_per_Job_Year ": parseFloat(document.getElementById("income_per_job").value),
                " Age": parseInt(document.getElementById("age").value),
                " Experience_per_House_Year": parseFloat(document.getElementById("exp_per_house").value),
                "Total_Years": parseInt(document.getElementById("total_years").value),
                "CURRENT_JOB_YRS ": parseInt(document.getElementById("current_job").value),
                "  Experience": parseInt(document.getElementById("experience").value),
                "Car_Ownership": document.getElementById("car_ownership").value,
                "Married/Single ": document.getElementById("marital_status").value,
                "House_Ownership": document.getElementById("house_ownership").value
            };

            fetch("/predict", {
                method: "POST",
                headers: { "Content-Type": "application/json" },
                body: JSON.stringify(data)
            })
            .then(response => response.json())
            .then(result => {
                document.getElementById("result").innerText = "Prediction: " + result.prediction;
            })
            .catch(error => {
                document.getElementById("result").innerText = "Error: " + error;
            });
        }
    </script>
</body>
</html>
"""

# Serve the frontend
@app.route("/")
def home():
    return render_template_string(HTML_PAGE)

# Define route for prediction
@app.route("/predict", methods=["POST"])
def predict():
    try:
        # Get JSON request data
        data = request.get_json()

        # Convert data to DataFrame
        df = pd.DataFrame([data])

        # Convert categorical values using mappings
        for col, mapping in category_mappings.items():
            if col in df.columns:
                df[col] = df[col].map(mapping)

        # Make prediction
        prediction = model.predict(df)[0]

        # Convert prediction to human-readable format
        result = "Risky" if prediction == 1 else "Not Risky"

        return jsonify({"prediction": result})
    except Exception as e:
        return jsonify({"error": str(e)})

# Function to run Flask in a separate thread
def run_flask():
    app.run(debug=True, use_reloader=False)

# Start Flask in a new thread
thread = threading.Thread(target=run_flask)
thread.start()
```
